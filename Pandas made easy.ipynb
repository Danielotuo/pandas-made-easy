## Pandas

Pandas is an open source library that allows you to explore, analyse and manipulate data. One main use of pandas is to transform our data for easier use with machine learning algorithms

```python
# Importing pandas
import pandas as pd
```

Pandas has two datatypes: Series and DataFrame. 
    - Series is a 1-d column of data
    - DataFframe is a 2-d data table with column and rows

```python
# Create a Series of countries
country = pd.Series(['Togo', 'Japan', 'Chile', 'Canada'])
country
```

```python
# Create a Series of Continents
continent = pd.Series(['Africa', 'Asia','South America', 'North America'])
continent
```

```python
  # Create a dataframe from the countries and continents
country_data = pd.DataFrame({'country':country,
                             'continent':continent})
country_data
```

### loc
It takes an integer as in input and retrieves the data index based on that integer

```python
# a sample series
cars = pd.Series(['tesla', 'toyota', 'benz', 'nissan'], index =[0,4,8,3])
cars
```

```python
# view index 8
cars.loc[8]
```

```python
# try with boston housing data
df.loc[45]
```

### iloc
iloc retrives exact index positions

```python
cars.iloc[3]
```

```python
cars
```

```python
# Select a particular column: use the column name
df['AGE']
```

```python
# select data with age over 80
df[df['AGE']>80]
```

```python
# pd.crosstab: two view columns together and compare them
# compare age and tax

pd.crosstab(df['AGE'], df['TAX'])
```

```python
# Use 'groupby' to compare a column in context with other columns
# Eg group ny age and the mean of the other columns

df.groupby(df['AGE']).mean()
```
